```python
# ============================
# Predictive Analytics Project
# Phase 1: Customer Behavior Analysis & EDA
# ============================

import pandas as pd
import numpy as np

# Create synthetic dataset
data = {
    "CustomerID": [101, 102, 103, 104, 105],
    "Age": [25, 34, 45, 29, 40],
    "Gender": ["Male", "Female", "Male", "Female", "Male"],
    "Income": [45000, 60000, 75000, 50000, 82000],
    "Spending": [1200, 2500, 3000, 1800, 4000],
    "Purchase_Date": ["2026-01-15", "2026-02-10", "2026-03-05", "2026-04-20", "2026-05-12"],
    "Segment": ["A", "B", "A", "C", "B"]
}

df = pd.DataFrame(data)

# ✅ Add Churn column (0 = Active, 1 = Churned)
df['Churn'] = np.random.choice([0, 1], size=len(df), p=[0.7, 0.3])

# Save dataset
df.to_csv("customer_data.csv", index=False)
print("✅ Synthetic dataset created with Churn column and saved as customer_data.csv")

# Quick checks
print(df)
print(df.info())
print(df.describe())
print(df.head())
print("Shape:", df.shape)   # rows, columns
print("Missing values:\n", df.isnull().sum())
```

```python
# ============================
# Step 3: Univariate Analysis
# ============================

import seaborn as sns
import matplotlib.pyplot as plt

# 1. Age Distribution
sns.histplot(df['Age'], bins=20, kde=True, color='skyblue')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
# 👉 Insight: Shows which age group dominates your customer base.

# 2. Income Distribution
sns.histplot(df['Income'], bins=20, kde=True, color='green')
plt.title("Income Distribution")
plt.xlabel("Income")
plt.ylabel("Count")
plt.show()
# 👉 Insight: Reveals spread of income levels (low vs high earners).

# 3. Spending Distribution
sns.histplot(df['Spending'], bins=20, kde=True, color='orange')
plt.title("Spending Distribution")
plt.xlabel("Spending")
plt.ylabel("Count")
plt.show()
# 👉 Insight: Identifies typical spending ranges and big spenders.

# Gender Distribution
sns.countplot(x='Gender', data=df, color='steelblue')
plt.title("Gender Distribution")
plt.show()

# Segment Distribution
sns.countplot(x='Segment', data=df, color='green')
plt.title("Customer Segment Distribution")
plt.show()

# Churn Distribution (only if column exists)
if 'Churn' in df.columns:
    sns.countplot(x='Churn', data=df, color='red')
    plt.title("Churn Distribution (0=Active, 1=Churned)")
    plt.show()
```

```python
# ============================
# Step 4: Bivariate Analysis
# ============================

import seaborn as sns
import matplotlib.pyplot as plt
# Income vs Spending (scatterplot with churn coloring)
if 'Churn' in df.columns:
    sns.scatterplot(x='Income', y='Spending', data=df, hue='Churn', palette='coolwarm')
    plt.title("Income vs Spending (colored by Churn)")
    plt.show()

# Income vs Churn
if 'Churn' in df.columns:
    sns.boxplot(x='Churn', y='Income', data=df)
    plt.title("Income vs Churn")
    plt.show()

# Spending vs Churn
if 'Churn' in df.columns:
    sns.boxplot(x='Churn', y='Spending', data=df)
    plt.title("Spending vs Churn")
    plt.show()

# Age vs Churn
if 'Churn' in df.columns:
    sns.boxplot(x='Churn', y='Age', data=df)
    plt.title("Age vs Churn")
    plt.show()

# Segment vs Churn
if 'Churn' in df.columns:
    sns.countplot(x='Segment', hue='Churn', data=df)
    plt.title("Churn by Segment")
    plt.show()

# Gender vs Churn
if 'Churn' in df.columns:
    sns.countplot(x='Gender', hue='Churn', data=df)
    plt.title("Churn by Gender")
    plt.show()
```
